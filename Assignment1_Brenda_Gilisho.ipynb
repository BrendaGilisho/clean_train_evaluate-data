```python
#Convert all columns with dates to be shown as dates when you run the data.info() instruction[5]
```

```python
columns = ['date_adm_icu_1', 'date_icu_disch_death_1', 'date_1',
           'date_of_admission_to_tbh', 'date_tbh_disch_death_1', 'dob', 'date']


# convert to datetime
for col in columns:
    data[col] = pd.to_datetime(data[col], format='%m/%d/%Y', errors='coerce')


print(data.info())
```

```python
# Drop all columns with dates[2]
```

```python
data.drop(['date_adm_icu_1', 'date_icu_disch_death_1', 'date_1',
           'date_of_admission_to_tbh', 'date_tbh_disch_death_1', 'dob', 'date'], axis = 1, inplace = True)

```

```python
# Encode all categorical variables except the Death column using the get_dummies() function[5]
```

```python
# Create a new dataframe with the categorical columns
df_categorical = data.select_dtypes(include=['object'])

# Create a new dataframe with the non-categorical columns
df_non_categorical = data.select_dtypes(exclude=['object'])

# Create a new dataframe with the categorical columns and imputed values
df_categorical_imputed = pd.get_dummies(df_categorical)

# Create a new dataframe with the non-categorical columns and imputed values
df_non_categorical_imputed = pd.DataFrame(df_non_categorical)

# Create a new dataframe with the imputed values
df_imputed = pd.concat(
    [df_categorical_imputed, df_non_categorical_imputed], axis=1)

# print the first 5 rows of the imputed data
print(df_imputed.head())
```

```python
# Encode the Death column using the integer encoding scheme[5]
```

```python
# Create a new dataframe with the categorical columns
df_categorical = data.select_dtypes(include=['object'])

# Create a new dataframe with the non-categorical columns
df_non_categorical = data.select_dtypes(exclude=['object'])

# Create a new dataframe with the categorical columns and imputed values
df_categorical_imputed = pd.DataFrame(df_categorical['discharge_death'].replace(
    {'Y': 1, 'N': 0}))

# Create a new dataframe with the non-categorical columns and imputed values
df_non_categorical_imputed = pd.DataFrame(df_non_categorical)

# Create a new dataframe with the imputed values
df_imputed = pd.concat(
    [df_categorical_imputed, df_non_categorical_imputed], axis=1)


# print the first 5 rows of the imputed data
print(df_imputed.head())
```

```python
# Scale all the numeric columns using the Standard Scaler[5]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Create a new dataframe with the numeric columns
df_numeric = data.select_dtypes(include=['int64', 'float64'])

# Create a new dataframe with the non-numeric columns
df_non_numeric = data.select_dtypes(exclude=['int64', 'float64'])

# Create a new dataframe with the numeric columns and imputed values
df_numeric_imputed = pd.DataFrame(StandardScaler().fit_transform(df_numeric))

# Create a new dataframe with the non-numeric columns and imputed values
df_non_numeric_imputed = pd.DataFrame(df_non_numeric)

# Create a new dataframe with the imputed values
df_imputed = pd.concat(
    [df_numeric_imputed, df_non_numeric_imputed], axis=1)

# print the first 5 rows of the imputed data
print(df_imputed.head())
```
